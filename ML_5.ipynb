```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("iris.csv")
```

```python
data.head()
```

```python
data.columns
```

```python
data.dtypes
```

```python
data["species"] = pd.Categorical(data["species"])
```

```python
data["sepal_length"] = data["sepal_length"].astype(np.float32)
```

```python
data["new_col"] = [int(x) for x in data["sepal_length"]]
```

```python
data = data.drop("new_col", axis=1)
```

```python
data.head()
```

```python
data.corr()
```

```python
sns.kdeplot(data["petal_length"])
plt.show()
```

```python
sns.scatterplot("sepal_length", "sepal_width", hue="species", data=data)
plt.show()
```

```python
sns.pairplot(data, hue="species")
plt.show()
```

```python
data.groupby("species").mean()
```
